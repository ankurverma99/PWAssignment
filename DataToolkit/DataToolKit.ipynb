```python
#How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np


arr = np.asarray([[1,2,3],
                  [4,5,6],
                  [7,8,9]])
print(arr)
sum = np.sum(arr,axis=1)
print("Sum of each Row",sum)

```

```python
#Write a Pandas script to find the mean of a specific column in a DataFrame

import pandas as pd

DataFrameA = pd.DataFrame({
    'Name': ['Ankur', 'Riya', 'Karan', 'Pooja'],
    'Age': [24, 22, 25, 28],
    'Salary': [50000, 60000, 55000, 42000]
})

mean_salary = DataFrameA['Salary'].mean()  
print(mean_salary)
```

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Age': [23, 25, 30, 22, 28],
    'Salary': [50000, 60000, 80000, 45000, 70000],
    'Experience': [2, 3, 8, 1, 5]
})


corr_matrix = data.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 20, 18]
y2 = [5, 7, 12, 14, 16]
y3 = [2, 3, 5, 7, 9]

# Create line plot with multiple lines
plt.plot(x, y1, label='Series 1', marker='o')
plt.plot(x, y2, label='Series 2', marker='s')
plt.plot(x, y3, label='Series 3', marker='^')

# Add title and labels
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
#Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd


data = {
    'Employee': ['Ankur', 'Riya', 'Karan', 'Pooja'],
    'Salary': [50000, 60000, 55000, 42000],
    'Age': [24, 22, 25, 28]
}

df = pd.DataFrame(data)


filtered_df = df[df['Salary'] > 50000]

print(filtered_df)
```

```python
#Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Salary': [50000, 60000, 55000, 42000, 70000, 65000, 48000, 53000]
})

sns.histplot(data['Salary'], bins=5, kde=True, color='skyblue')


plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
#Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Employee': ['Ankur', 'Riya', 'Karan', 'Pooja', 'Vikram'],
    'Salary': [50000, 60000, 55000, 42000, 70000],
    'Age': [24, 22, 25, 28, 30],
    'Experience': [2, 3, 5, 4, 8]
})

# Create 3D scatter plot
fig = px.scatter_3d(
    data,
    x='Age',
    y='Salary',
    z='Experience',
    color='Salary',
    size='Experience',
    hover_name='Employee',
    title='3D Scatter Plot of Employees'
)

# Display plot
fig.show()
```
